```python
# Importa a biblioteca
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# URL do arquivo CSV no repositório GitHub (conteúdo 'raw')
github_url = 'https://raw.githubusercontent.com/ElenirOliveira/churn-insight-api/main/WA_Fn-UseC_-Telco-Customer-Churn.csv'

try:
    # Carrega os dados do GitHub em um DataFrame do pandas
    dados_github = pd.read_csv(github_url)
    print("Dados carregados com sucesso do GitHub!")
    print(dados_github.head())
except Exception as e:
    print(f"Erro ao carregar dados do GitHub: {e}")
    print("Verifique se a URL está correta e se o arquivo CSV está acessível.")
```

```python
columns_traducao = {
    'customerID': 'ID_Cliente',
    'gender': 'Gênero',
    'SeniorCitizen': 'Cidadão_Sênior',
    'Partner': 'Parceiro',
    'Dependents': 'Dependentes',
    'tenure': 'Permanência_Meses',
    'PhoneService': 'Serviço_Telefone',
    'MultipleLines': 'Multiplas_Linhas',
    'InternetService': 'Serviço_Internet',
    'OnlineSecurity': 'Segurança_Online',
    'OnlineBackup': 'Backup_Online',
    'DeviceProtection': 'Proteção_Dispositivo',
    'TechSupport': 'Suporte_Técnico',
    'StreamingTV': 'TV_Streaming',
    'StreamingMovies': 'Filmes_Streaming',
    'Contract': 'Contrato',
    'PaperlessBilling': 'Fatura_Sem_Papel',
    'PaymentMethod': 'Método_Pagamento',
    'MonthlyCharges': 'Cobranças_Mensais',
    'TotalCharges': 'Cobranças_Totais',
    'Churn': 'Evasão'
}

dados_github =  dados_github.rename(columns=columns_traducao)
print("Colunas traduzidas com sucesso!")
print(dados_github.columns)
```

```python
dados_github
```

```python
translation_map = {
    'Female': 'Feminino',
    'Male': 'Masculino',
    'Yes': 'Sim',
    'No': 'Não',
    'No phone service': 'Sem serviço de telefone',
    'DSL': 'DSL',
    'Fiber optic': 'Fibra ótica',
    'No internet service': 'Sem serviço de internet',
    'Month-to-month': 'Mensal',
    'One year': 'Um ano',
    'Two year': 'Dois anos',
    'Electronic check': 'Cheque eletrônico',
    'Mailed check': 'Cheque enviado',
    'Bank transfer (automatic)': 'Transferência bancária (automática)',
    'Credit card (automatic)': 'Cartão de crédito (automático)'
}

for column in  dados_github.select_dtypes(include='object').columns:
    if column != 'ID_Cliente':
         dados_github[column] = dados_github[column].replace(translation_map)

print("Valores categóricos traduzidos com sucesso!")
```

```python
print( dados_github['Cobranças_Totais'].unique())
dados_github['Cobranças_Totais'] = pd.to_numeric( dados_github['Cobranças_Totais'], errors='coerce')
print(f"Valores nulos introduzidos na coluna 'Cobranças_Totais': { dados_github['Cobranças_Totais'].isnull().sum()}")
```

```python
for column in  dados_github.select_dtypes(include='object').columns:
    if column != 'ID_Cliente': # ID_Cliente usually doesn't need translation
        print(f"\nColuna: {column}")
        print( dados_github[column].unique())
```
